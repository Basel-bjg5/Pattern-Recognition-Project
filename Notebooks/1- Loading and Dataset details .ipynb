```python
import os
import matplotlib.pyplot as plt
import numpy as np

base_path = r"\\?\C:\Users\pc\Documents\Basel BME\BME@FEHU\3rd BME first term\2) Pattern Recognition\project\Project - Kidney Stone Detection\Axial CT Imaging Dataset for AI-Powered Kidney Stone Detection A Resource for Deep Learning Research\Axial CT Imaging Dataset for AI-Powered Kidney Stone Detection A Resource for Deep Learning Research\Kindy Stone Dataset"
st_path = os.path.join(base_path , "Original Dataset") #path of original dataset
nd_path = os.path.join(base_path , "Augmented Dataset") #path of agumented dataset

clusters   = [folder for folder in os.listdir(st_path) ]
A_closters = [folder for folder in os.listdir(nd_path) ]

print(f"For Original dataset :\nNumber of clusters (classes): {len(clusters)}\nCluster names:{clusters} \n")
print(f"For Augmented dataset :\nNumber of clusters (classes): {len(A_closters)}\nCluster names:{A_closters} \n")
```

```python
img_exts = {'.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff'}

# --- Original Dataset ---
Number_of_pics = []
counts_by_folder = {}

for folder in clusters:
    folder_path = os.path.join(st_path, folder)
    files = [
        f for f in os.listdir(folder_path) if os.path.splitext(f)[1].lower() in img_exts
    ]
    Number_of_pics.append(len(files))
    counts_by_folder[folder] = len(files)

print('\nüìÅ Counts per folder for **Original dataset**:')
for name, cnt in zip(clusters, Number_of_pics):
    print(f'  {name}: {cnt}')


# --- Augmented Dataset ---
O_Number_of_pics = []
O_counts_by_folder = {}

for folder in A_closters:
    folder_path = os.path.join(nd_path, folder)
    files = [
        f for f in os.listdir(folder_path) if os.path.splitext(f)[1].lower() in img_exts
    ]
    O_Number_of_pics.append(len(files))
    O_counts_by_folder[folder] = len(files)

print('\nüìÅ Counts per folder for **Augmented dataset**:')
for name, cnt in zip(A_closters, O_Number_of_pics):
    print(f'  {name}: {cnt}')


# --- Bar Chart ---
x = np.arange(len(clusters))
width = 0.5

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width/2, Number_of_pics, width, label='Original', color='skyblue')
ax.bar(x + width/2, O_Number_of_pics, width, label='Augmented', color='orange')

ax.set_xticks(x)
ax.set_xticklabels(clusters, rotation=45)
ax.set_ylabel('Number of Images')
ax.set_title('Image Counts per Class: Original vs Augmented')
ax.legend()

plt.tight_layout()
plt.show()
```
